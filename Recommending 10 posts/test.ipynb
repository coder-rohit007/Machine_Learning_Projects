```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_json('Sample_dataset_Campus24_Global_posts_json.json')
```

```python
#raw data
df.head()
```

```python
#melting and creating new dataframe

comments = []
images = []
likes = []
tags = []
texts = []

for i in range(92):
    comments.append(df.iloc[0][i])
    images.append(df.iloc[1][i])
    likes.append(df.iloc[2][i])
    tags.append(df.iloc[3][i])
    texts.append(df.iloc[4][i])
    
new_df = pd.DataFrame({'timestamp': df.columns, 'comments':comments, 'likes':likes, 'tags':tags, 'texts':texts})
```

```python
data = new_df.copy()
```

```python
#filling missing values

data['texts'] = data['texts'].fillna('')
data = data.fillna(0)
```

```python
num_comments = []
num_likes = []
for i in range(data.shape[0]):
    if data.loc[i]['comments'] == 0:
        num_comments.append(0)
    else:
        num_comments.append(len(data.loc[i]['comments']))
    if data.loc[i]['likes'] == 0:
        num_likes.append(0)
    else:
        num_likes.append(len(data.loc[i]['likes']))
        
data['num_likes'] = pd.Series(num_likes)
data['num_comments'] = pd.Series(num_comments)
```

```python
#allocates rank to each post
def alloting_rank(df):
    df['rank'] = 2*df['num_likes'] + 3*df['num_comments']
    return df

#return the first 10 posts to show
def first_10_posts(df):
    df = alloting_rank(df)
    df = df.sort_values('rank', ascending=False)
    df = df.head(10)
    return df
```
