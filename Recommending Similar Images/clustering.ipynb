```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
IMG_SIZE = 70
path = r"C:\Users\PRATIK\Desktop\Groomup\Cosine check\diamond"

arrays = []
img_names = []
for img in os.listdir(path):
    img1 = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)  #reading in one image at a time
    new_array1 = cv2.resize(img1, (IMG_SIZE, IMG_SIZE))
    length = np.prod(new_array1.shape)
    new_array1 = new_array1.reshape(length)
    img_names.append(img)
    arrays.append(new_array1)
    
X = np.array(arrays)
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
labels = kmeans.labels_
```

```python
zeros = []
ones = []
twos = []
for i in range(len(img_names)):
    if labels[i]==0:
        zeros.append(img_names[i])
    elif labels[i]==1:
        ones.append(img_names[i])
    else:
        twos.append(img_names[i])
```

```python
for img in os.listdir(path):
    if img in zeros:
        img1 = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img1)
        plt.show()
```

```python
for img in os.listdir(path):
    if img in ones:
        img1 = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img1)
        plt.show()
```

```python
for img in os.listdir(path):
    if img in twos:
        img1 = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img1)
        plt.show()
```
